```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_Counts'})
```

```python
ratings
```

##Observation:
1.when rating in between 4.5 to 4.9---> Excellent	
2.when rating in between 4.0 to 4.4--->Very Good
3.when rating in between 3.5 to 3.9--->Good
4.when rating in between 2.5 to 3.4--->Average
5.when rating in between 1.8 to 2.4--->Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,3)
```

```python
sns.barplot(x='Aggregate rating',y='Rating_Counts',hue='Rating color',data=ratings,palette = ['#FFFFFF', '#FF0000', '#FFA500', '#FFFF00', '#008000', '#006400'])
```

##Observation:
1.Not rated count is high
2.maximum number of rating between 2.5 to 3.4

```python
palette = ['#FFFFFF', '#FF0000', '#FFA500', '#FFFF00', '#008000', '#006400']
sns.countplot(x="Rating color",data=ratings,palette=palette)
```
